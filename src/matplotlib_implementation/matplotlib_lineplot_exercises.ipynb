```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Write a Python program to draw a line with suitable label in the x axis, y axis and a title
x = np.array([20, 30, 40, 50, 60])
y = np.array([10, 50, 30, 60, 10])
plt.plot(x,y, 'o:g')
plt.xlabel("X-axis")
plt.ylabel('Y-axis')
plt.title("Line plot")
plt.show()
```

```python
# 4. Write a Python program to draw line charts of the financial data of Alphabet Inc. between
# October 3, 2016 to October 7, 2016.
data = pd.read_csv('fdata.csv')
plt.plot(data['Date'],data['Open'], 'o--g', label = 'Open')
plt.plot(data['Date'],data['Close'], 'o--r', label = 'Close')
plt.plot(data['Date'],data['High'],'o--y', label = 'High')
plt.plot(data['Date'],data['Low'],'o--b', label = 'Low')
plt.xlabel('Date')
plt.ylabel('Prices')
plt.legend()
plt.title("Financial data of Alphabet Inc.")
plt.show()
```

```python
# 5. Write a Python program to plot two or more lines on same plot with suitable legends of each line.
data = {
    "Subjects":["English","Kannada","Hindi","Mathematics","Science","Social Studies"],
    "Student1":[87,78,56,89,67,78],
    "Student2":[93,88,78,72,90,87],
    "Student3":[82,90,56,35,89,96],
    "Student4":[38,44,35,67,40,54],
    "Student5":[54,90,67,34,67,50]
}

plt.figure(figsize=(12,8))
plt.plot(data["Subjects"],data["Student1"], 'o--r', label="Student 1")
plt.plot(data["Subjects"],data["Student2"], 'o-.g', label="Student 2")
plt.plot(data["Subjects"],data["Student3"], 'o:b', label="Student 3")
plt.plot(data["Subjects"],data["Student4"], 'o-y', label="Student 4")
plt.plot(data["Subjects"],data["Student5"], 'o--k', label="Student 5")
plt.xlabel("Subjects", fontsize = 16)
plt.ylabel("Marks", fontsize = 16)
plt.title("Marks scored by students", fontsize = 20, fontweight = 'bold')
plt.legend(loc = 9, title = 'Students') # Upper center
plt.grid(linestyle = ':')
plt.show()
```

```python
# 12. Write a Python program to create multiple types of charts.
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.figure(figsize=(15,12))

plt.subplot(2,2,1)
plt.plot(x,y, 'o-')
plt.title("Line Chart")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()

plt.subplot(2,2,2)
plt.bar(x,y, zorder = 3)
plt.title("Bar Chart")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(axis='y')

plt.subplot(2,2,3)
plt.pie(y , labels = x, autopct = '%1.1f%%')
plt.title("Pie Chart")

plt.subplot(2,2,4)
plt.scatter(x,y)
plt.title("Scatter Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```

```python
# 13. Write a Python program to display the grid and draw line charts of the closing value of
# Alphabet Inc. between October 3, 2016 to October 7, 2016. Customized the grid lines with
# linestyle -, width .5. and color blue.

data = pd.read_csv("fdata.csv")
plt.plot(data["Date"], data["Close"], marker = 'o', linestyle = '-', linewidth = 0.5, color = 'blue')
plt.xlabel("Date")
plt.ylabel("Closing value")
plt.grid(linestyle = '--')
plt.show()
```

```python
# 14. Write a Python program to display the grid and draw line charts of the closing value of Alphabet Inc.
# between October 3, 2016 to October 7, 2016. Customized the grid lines with rendering with a 
# larger grid (major grid) and a smaller grid (minor grid).Turn on the grid but turn off ticks
plt.plot(data['Date'],data['Close'],marker="o",label="closed price")

plt.xlabel("Date")
plt.ylabel("Close")
plt.title("Alphabet Inc between October 3, 2016 to October 7, 2016")

plt.legend()

plt.grid(which='major',ls="-",lw=0.5,color='blue')
plt.minorticks_on()
plt.grid(which='minor',ls='--',lw=0.5,color="blue")
plt.tick_params(axis='both', which='both', length=0)

plt.show()
```
